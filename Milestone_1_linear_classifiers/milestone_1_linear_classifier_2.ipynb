```python
def train_val_split(X, y, val_ratio=0.2, random_seed=None):

    if random_seed is not None:
        np.random.seed(random_seed)
    
    data_size = X.shape[0]

    # Generate a permutation of indices
    indices = np.random.permutation(data_size)

    # Compute the split point
    split_point = int((1 - val_ratio) * data_size)

    # Split indices for training and validation
    train_indices = indices[:split_point]
    val_indices = indices[split_point:]

    # Use indices to split the data
    X_train = X[train_indices]
    y_train = y[train_indices]
    X_val = X[val_indices]
    y_val = y[val_indices]

    return X_train, y_train, X_val, y_val
```

```python
def undersample_data(X, y):
    # Identify unique classes and their counts
    unique_classes, class_counts = np.unique(y, return_counts=True)
    print(f"Class counts before undersampling: {dict(zip(unique_classes, class_counts))}")
    
    # Find the smallest class size
    min_class_size = np.min(class_counts)
    
    # Initialize lists to store undersampled data
    X_undersampled = []
    y_undersampled = []
    
    # Iterate through each class and undersample
    for cls in unique_classes:
        # Get indices of the current class
        class_indices = np.where(y == cls)[0]
        
        # Randomly select samples equal to the minimum class size
        undersampled_indices = np.random.choice(class_indices, size=min_class_size, replace=False)
        
        # Append the undersampled data
        X_undersampled.append(X[undersampled_indices])
        y_undersampled.append(y[undersampled_indices])
    
    # Combine all undersampled data
    X_undersampled = np.vstack(X_undersampled)
    y_undersampled = np.concatenate(y_undersampled)
    
    # Shuffle the undersampled data
    shuffled_indices = np.random.permutation(len(y_undersampled))
    X_undersampled = X_undersampled[shuffled_indices]
    y_undersampled = y_undersampled[shuffled_indices]
    
    print(f"Class counts after undersampling: {np.unique(y_undersampled, return_counts=True)}")
    
    return X_undersampled, y_undersampled
```

```python
def normalize_features(data):
    """
    Normalize features (mean = 0, variance = 1).
    """
    X = data
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    X_normalized = (X - mean) / (std + 1e-8)  # Add small epsilon to avoid division by zero
    return X_normalized
```

```python
class MulticlassLinearModel:
    """
    Abstract class for all multiclass linear models.
    """

    def __init__(self, w0, reg, num_classes):
        """
        Initialize model weights and biases.
        """
        self.w = np.array(w0, dtype=float)  # Initialize weights
        self.reg = reg  # Regularization hyperparameter
        self.num_classes = num_classes  # Number of classes

    def predict_logits(self, X):
        """
        Return logits for a batch X.
        X: n_examples x n_features
        Returns: n_examples x num_classes (logits for each class)
        """
        return np.dot(X, self.w)

    def predict(self, X):
        """
        Return predicted class labels for a batch X.
        X: n_examples x n_features
        Returns: n_examples (predicted class labels)
        """
        logits = self.predict_logits(X)
        return np.argmax(logits, axis=1)

    def error_rate(self, X, y):
        """
        Compute the multiclass error rate.
        """
        preds = self.predict(X)
        return np.mean(preds != y)

    def loss(self, X, y):
        """
        Placeholder for computing the loss.
        """
        return 0

    def gradient(self, X, y):
        """
        Placeholder for computing the gradient.
        """
        return self.w

    def train(self, X, y, stepsize, n_steps, plot=False):
        """
        Train with gradient descent for multiclass classification.

        Parameters:
            X (ndarray): Training features (num_samples x num_features).
            y (ndarray): Training labels (num_samples,).
            stepsize (float): Gradient descent step size.
            n_steps (int): Number of training steps.
            plot (bool): Whether to plot loss during training.

        Returns:
            tuple: Arrays of losses and errors over training steps.
        """
        losses = []
        errors = []

        for i in range(n_steps):
            # Gradient Descent
            self.w -= stepsize * self.gradient(X, y)

            # Update losses
            losses.append(self.loss(X, y))

            # Update errors
            errors.append(self.error_rate(X, y))

        print(f"Training completed: the train error is {errors[-1]*100:.2f}%")

        # Optionally plot loss
        if plot:
            plt.plot(range(n_steps), losses)
            plt.title("Training Loss")
            plt.xlabel("Step")
            plt.ylabel("Loss")
            plt.show()

        return np.array(losses), np.array(errors)
```

```python
class MulticlassLinearRegression(MulticlassLinearModel):
    def loss(self, X, y):
        y_one_hot = np.eye(self.num_classes)[y]  # One-hot encoding
        preds = self.predict_logits(X)
        return 0.5 * np.mean((preds - y_one_hot) ** 2) + 0.5 * self.reg * np.sum(self.w ** 2)

    def gradient(self, X, y):
        y_one_hot = np.eye(self.num_classes)[y]
        preds = self.predict_logits(X)
        grad = np.dot(X.T, (preds - y_one_hot)) / X.shape[0]
        return grad + self.reg * self.w
```

```python
class MulticlassPerceptron(MulticlassLinearModel):
    def loss(self, X, y):
        preds = self.predict_logits(X)
        y_one_hot = np.eye(self.num_classes)[y]
        return np.mean(np.maximum(0, 1 - preds * y_one_hot)) + 0.5 * self.reg * np.sum(self.w ** 2)

    def gradient(self, X, y):
        y_one_hot = np.eye(self.num_classes)[y]
        preds = self.predict_logits(X)
        active = (1 - preds * y_one_hot > 0).astype(float)
        grad = np.dot(X.T, -active * y_one_hot) / X.shape[0]
        return grad + self.reg * self.w
```

```python
class MulticlassSVM(MulticlassLinearModel):
    def loss(self, X, y):
        preds = self.predict_logits(X)
        correct_class_scores = preds[np.arange(X.shape[0]), y][:, np.newaxis]
        margins = np.maximum(0, preds - correct_class_scores + 1)
        margins[np.arange(X.shape[0]), y] = 0
        return np.mean(np.sum(margins, axis=1)) + 0.5 * self.reg * np.sum(self.w ** 2)

    def gradient(self, X, y):
        preds = self.predict_logits(X)
        correct_class_scores = preds[np.arange(X.shape[0]), y][:, np.newaxis]
        margins = preds - correct_class_scores + 1
        margins[margins < 0] = 0
        margins[np.arange(X.shape[0]), y] = 0

        coeff = (margins > 0).astype(float)
        coeff[np.arange(X.shape[0]), y] -= np.sum(coeff, axis=1)
        grad = np.dot(X.T, coeff) / X.shape[0]
        return grad + self.reg * self.w
```

```python
class MulticlassLogisticRegression(MulticlassLinearModel):
    def loss(self, X, y):
        preds = self.predict_logits(X)
        preds = np.exp(preds - np.max(preds, axis=1, keepdims=True))
        probs = preds / np.sum(preds, axis=1, keepdims=True)
        correct_class_probs = probs[np.arange(X.shape[0]), y]
        return -np.mean(np.log(correct_class_probs)) + 0.5 * self.reg * np.sum(self.w ** 2)

    def gradient(self, X, y):
        preds = self.predict_logits(X)
        preds = np.exp(preds - np.max(preds))
        probs = preds / np.sum(preds, axis=1, keepdims=True)
        probs[np.arange(X.shape[0]), y] -= 1
        grad = np.dot(X.T, probs) / X.shape[0]
        return grad + self.reg * self.w
```

```python
def calculate_accuracy(y_true, y_pred):
    """
    Calculate accuracy for classification.

    Parameters:
        y_true (numpy.ndarray): Ground truth labels.
        y_pred (numpy.ndarray): Predicted labels.

    Returns:
        float: Accuracy as a proportion (0 to 1).
    """
    correct_predictions = np.sum(y_true == y_pred)
    total_predictions = y_true.shape[0]
    accuracy = correct_predictions / total_predictions
    return accuracy
```

```python
def compute_feature_importance(model, num_features, num_classes):
    """
    Compute feature importance based on the learned weights of a model.

    Parameters:
        model: Trained model with a `w` attribute containing the weights.
        num_features: Number of features.
        num_classes: Number of classes.

    Returns:
        feature_importance: Importance scores for each feature.
    """
    # Get absolute weights for each feature across all classes
    weights = np.abs(model.w)
    feature_importance = np.mean(weights, axis=1)  # Average importance across classes
    return feature_importance


def select_top_features(feature_importance, top_k):
    """
    Select the indices of the top-k most important features.

    Parameters:
        feature_importance: Importance scores for each feature.
        top_k: Number of top features to select.

    Returns:
        indices of top-k features.
    """
    return np.argsort(feature_importance)[-top_k:]




```

```python
def evaluate_models_and_select_features(models, X_train, y_train, X_test, y_test, num_features, num_classes, top_k=10, stepsize=0.1, n_steps=100):
    """
    Train and evaluate models to select top-k features and evaluate test set performance.

    Parameters:
        models (dict): Dictionary of model names and their classes.
        X_train (ndarray): Training features.
        y_train (ndarray): Training labels.
        X_test (ndarray): Test features.
        y_test (ndarray): Test labels.
        num_features (int): Number of features in the dataset.
        num_classes (int): Number of classes in the dataset.
        top_k (int): Number of top features to select.
        stepsize (float): Step size for gradient descent.
        n_steps (int): Number of training steps.

    Returns:
        dict: Top-k selected features for each model.
    """
    # Dictionary to store selected features
    selected_features = {}

    for model_name, model_class in models.items():
        print(f"Evaluating {model_name}...")
        
        # Initialize model weights
        w0 = np.random.randn(num_features, num_classes)
        
        # Initialize and train the model
        model = model_class(w0, reg=0.1, num_classes=num_classes)
        train_losses, train_errors = model.train(X_train, y_train, stepsize, n_steps)
    
        # Compute feature importance
        feature_importance = compute_feature_importance(model, num_features, num_classes)
    
        # Select top-k features
        top_features = select_top_features(feature_importance, top_k)
        selected_features[model_name] = top_features
    
        # Print and plot feature importance
        print(f"Top-{top_k} features for {model_name}: {top_features}")
        plt.figure(figsize=(8, 4))
        plt.bar(range(num_features), feature_importance)
        plt.title(f"Feature Importance - {model_name}")
        plt.xlabel("Feature Index")
        plt.ylabel("Importance")
        plt.show()
    
        # Evaluate on test set
        logits = model.predict_logits(X_test)  # Get raw logits
        predictions = model.predict(X_test)   # Get class predictions

        # Display raw logits for debugging
        print(f"Logits for {model_name}:\n{logits[:5]}")  # Show first 5 examples
        accuracy = calculate_accuracy(y_test, predictions)
        print(f"Test Accuracy for {model_name}: {accuracy:.4f}")
    
        # Plot class distribution of predictions
        plot_class_distribution(predictions)
    
    return selected_features
```

```python
import numpy as np
import matplotlib.pyplot as plt

def evaluate_models_and_plot_performance(models, X_train, y_train, X_test, y_test, num_features, num_classes, stepsize=0.1, n_steps=100):
    """
    Train and evaluate models, plotting loss and accuracy over training steps.

    Parameters:
        models (dict): Dictionary of model names and their classes.
        X_train (ndarray): Training features.
        y_train (ndarray): Training labels.
        X_test (ndarray): Test features.
        y_test (ndarray): Test labels.
        num_features (int): Number of features in the dataset.
        num_classes (int): Number of classes in the dataset.
        stepsize (float): Step size for gradient descent.
        n_steps (int): Number of training steps.

    Returns:
        dict: Dictionary with train and test performance metrics for each model.
    """
    performance_metrics = {}

    for model_name, model_class in models.items():
        print(f"Training and evaluating {model_name}...")

        # Initialize model weights
        w0 = np.random.randn(num_features, num_classes)

        # Initialize the model
        model = model_class(w0, reg=0.1, num_classes=num_classes)

        train_losses = []
        train_accuracies = []
        test_losses = []
        test_accuracies = []

        # Training loop
        for step in range(n_steps):
            # Perform a single step of training
            train_loss, _ = model.train(X_train, y_train, stepsize, n_steps=1)
            train_losses.append(train_loss[-1])
            
            # Evaluate on training set
            train_accuracy = 1 - model.error_rate(X_train, y_train)
            train_accuracies.append(train_accuracy)

            # Evaluate on test set
            test_loss = model.loss(X_test, y_test)
            test_accuracy = 1 - model.error_rate(X_test, y_test)
            test_losses.append(test_loss)
            test_accuracies.append(test_accuracy)

            # Print progress every 10 steps
            if step % 10 == 0 or step == n_steps - 1:
                print(f"Step {step}/{n_steps}: "
                      f"Train Loss={train_loss[-1]:.4f}, Train Acc={train_accuracy:.4f}, "
                      f"Test Loss={test_loss:.4f}, Test Acc={test_accuracy:.4f}")

                 # Generate predictions on test set
                predictions = model.predict(X_test)

                # Plot class distribution of predictions
                print(f"Class Distribution of Predictions at Step {step}:")
                plot_class_distribution(predictions)

        # Plot training and test loss
        plt.figure(figsize=(12, 6))
        plt.plot(range(n_steps), train_losses, label='Train Loss')
        plt.plot(range(n_steps), test_losses, label='Test Loss')
        plt.title(f'Loss Curve - {model_name}')
        plt.xlabel('Training Steps')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()

        # Plot training and test accuracy
        plt.figure(figsize=(12, 6))
        plt.plot(range(n_steps), train_accuracies, label='Train Accuracy')
        plt.plot(range(n_steps), test_accuracies, label='Test Accuracy')
        plt.title(f'Accuracy Curve - {model_name}')
        plt.xlabel('Training Steps')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.show()

        # Save metrics for this model
        performance_metrics[model_name] = {
            "train_losses": train_losses,
            "train_accuracies": train_accuracies,
            "test_losses": test_losses,
            "test_accuracies": test_accuracies
        }

    return performance_metrics
```
