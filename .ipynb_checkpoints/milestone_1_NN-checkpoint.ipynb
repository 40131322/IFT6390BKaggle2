# Split to validation set

```python
def train_val_split(X, y, val_ratio=0.2, random_seed=None):

    if random_seed is not None:
        np.random.seed(random_seed)
    
    data_size = X.shape[0]

    # Generate a permutation of indices
    indices = np.random.permutation(data_size)

    # Compute the split point
    split_point = int((1 - val_ratio) * data_size)

    # Split indices for training and validation
    train_indices = indices[:split_point]
    val_indices = indices[split_point:]

    # Use indices to split the data
    X_train = X[train_indices]
    y_train = y[train_indices]
    X_val = X[val_indices]
    y_val = y[val_indices]

    return X_train, y_train, X_val, y_val
```

```python
def calculate_pixel_label_correlation(X, y):
    """
    Calculate the correlation between each pixel value and the classification label.

    Parameters:
    - X: numpy array of shape (N, 28, 28), where N is the number of samples.
    - y: numpy array of shape (N,), classification labels.

    Returns:
    - correlation_grid: numpy array of shape (28, 28), containing the correlation for each pixel.
    """
    # Flatten the images into vectors of shape (N, 784)
    X_flat = X.reshape(X.shape[0], -1)  # Shape becomes (N, 784)
    
    # Initialize a correlation grid
    correlation_grid = np.zeros((28, 28))
    
    # Calculate the correlation for each pixel
    for i in range(28):
        for j in range(28):
            # Extract the pixel values at position (i, j) across all samples
            pixel_values = X[:, i, j]
            
            # Compute the correlation coefficient between pixel values and labels
            correlation = np.corrcoef(pixel_values, y)[0, 1]
            correlation_grid[i, j] = correlation
    
    return correlation_grid
corr = calculate_pixel_label_correlation(X,y)
print(corr.shape)
```

# Convuluted Neural Network

```python
def conv2d(X, kernel):
    """
    Perform a 2D convolution.
    
    Parameters:
        X: Input array (H, W)
        kernel: Kernel array (kH, kW)
        
    Returns:
        Output array after convolution.
    """
    H, W = X.shape
    kH, kW = kernel.shape
    output_h, output_w = H - kH + 1, W - kW + 1
    
    output = np.zeros((output_h, output_w))
    for i in range(output_h):
        for j in range(output_w):
            region = X[i:i+kH, j:j+kW]
            output[i, j] = np.sum(region * kernel)
    return output
```

```python
def max_pooling(X, pool_size=2, stride=2):
    """
    Perform max pooling.
    
    Parameters:
        X: Input array (H, W)
        pool_size: Size of the pooling window.
        stride: Stride size.
        
    Returns:
        Downsampled output.
    """
    H, W = X.shape
    output_h = (H - pool_size) // stride + 1
    output_w = (W - pool_size) // stride + 1
    
    output = np.zeros((output_h, output_w))
    for i in range(0, H - pool_size + 1, stride):
        for j in range(0, W - pool_size + 1, stride):
            region = X[i:i+pool_size, j:j+pool_size]
            output[i//stride, j//stride] = np.max(region)
    return output
```

```python
def relu(X):
    """
    Apply ReLU activation function.
    """
    return np.maximum(0, X)
```

```python
def flatten(X):
    """
    Flatten a 2D array into 1D.
    """
    return X.flatten()
```

```python
def dense(X, weights, biases):
    """
    Fully connected layer.
    
    Parameters:
        X: Input array (n_features,)
        weights: Weight matrix (n_features, n_outputs)
        biases: Bias vector (n_outputs,)
        
    Returns:
        Output after linear transformation.
    """
    return np.dot(X, weights) + biases
```

```python
def softmax(X):
    """
    Apply softmax function to an array.
    """
    exp_X = np.exp(X - np.max(X))  # Stabilize by subtracting max
    return exp_X / np.sum(exp_X)
```

```python
def forward_pass(X, params):
    """
    Perform a forward pass through a simple CNN.
    
    Parameters:
        X: Input image (H, W)
        params: Dictionary of weights and biases.
        
    Returns:
        Output probabilities.
    """
 # Convolution + ReLU
    conv1 = relu(conv2d(X, params['conv1_kernel']))
    print(f"Conv1 Output Shape: {conv1.shape}")
    
    # Max Pooling
    pool1 = max_pooling(conv1)
    print(f"Pool1 Output Shape: {pool1.shape}")
    
    # Flatten
    flat = flatten(pool1)
    print(f"Flatten Output Shape: {flat.shape}")
    
    # Fully Connected + ReLU
    fc1 = relu(dense(flat, params['fc1_weights'], params['fc1_biases']))
    
    # Output Layer (Softmax)
    output = softmax(dense(fc1, params['fc2_weights'], params['fc2_biases']))
    
    return output
```

```python
def dense_backward(d_out, input, weights):
    """
    Backward pass for a fully connected (dense) layer.
    
    Parameters:
        d_out: Gradient of the loss with respect to the output (n_outputs,).
        input: Input to the dense layer (n_features,).
        weights: Weights of the dense layer (n_features, n_outputs).
    
    Returns:
        d_input: Gradient of the loss with respect to the input.
        d_weights: Gradient of the loss with respect to the weights.
        d_biases: Gradient of the loss with respect to the biases.
    """
    d_input = np.dot(d_out, weights.T)
    d_weights = np.outer(input, d_out)
    d_biases = d_out
    return d_input, d_weights, d_biases
```

```python
def relu_backward(d_out, input):
    """
    Backward pass for ReLU activation.
    
    Parameters:
        d_out: Gradient of the loss with respect to the output.
        input: Input to the ReLU function.
    
    Returns:
        d_input: Gradient of the loss with respect to the input.
    """
    d_input = d_out * (input > 0)  # Gradient is 1 where input > 0, otherwise 0
    return d_input
```

```python
def max_pooling_backward(d_out, input, pool_size=2, stride=2):
    """
    Backward pass for max pooling.
    
    Parameters:
        d_out: Gradient of the loss with respect to the pooled output.
        input: Input to the pooling layer.
        pool_size: Size of the pooling window.
        stride: Stride size.
    
    Returns:
        d_input: Gradient of the loss with respect to the input.
    """
    H, W = input.shape
    d_input = np.zeros_like(input)
    output_h, output_w = d_out.shape

    for i in range(output_h):
        for j in range(output_w):
            # Get the pooling region
            pool_region = input[i*stride:i*stride+pool_size, j*stride:j*stride+pool_size]
            max_val = np.max(pool_region)
            
            # Distribute the gradient only to the max value location
            for m in range(pool_size):
                for n in range(pool_size):
                    if pool_region[m, n] == max_val:
                        d_input[i*stride+m, j*stride+n] = d_out[i, j]
    return d_input
```

```python
def conv2d_backward(d_out, input, kernel):
    """
    Backward pass for a convolutional layer.
    
    Parameters:
        d_out: Gradient of the loss with respect to the output (H_out, W_out).
        input: Input to the convolutional layer (H, W).
        kernel: Convolutional kernel (kH, kW).
    
    Returns:
        d_input: Gradient of the loss with respect to the input.
        d_kernel: Gradient of the loss with respect to the kernel.
    """
    H, W = input.shape
    kH, kW = kernel.shape
    d_input = np.zeros_like(input)
    d_kernel = np.zeros_like(kernel)
    
    H_out, W_out = d_out.shape
    for i in range(H_out):
        for j in range(W_out):
            region = input[i:i+kH, j:j+kW]
            d_kernel += region * d_out[i, j]
            d_input[i:i+kH, j:j+kW] += kernel * d_out[i, j]
    return d_input, d_kernel
```

```python
def backprop(X, y, output, params, cache):
    """
    Perform backpropagation for a simple CNN.
    
    Parameters:
        X: Input image (H, W).
        y: True label (scalar).
        output: Output probabilities from the forward pass.
        params: Dictionary of weights and biases.
        cache: Intermediate values from the forward pass.
    
    Returns:
        grads: Gradients for all parameters in the model.
    """
    grads = {}

    # Output layer (Softmax + Cross-Entropy Loss)
    d_out = output
    d_out[y] -= 1  # Gradient of loss with respect to softmax output

    # Dense Layer 2 (Output Layer)
    d_fc2, grads['fc2_weights'], grads['fc2_biases'] = dense_backward(
        d_out, cache['fc1'], params['fc2_weights']
    )

    # Dense Layer 1
    d_fc1 = relu_backward(d_fc2, cache['fc1'])
    d_fc1, grads['fc1_weights'], grads['fc1_biases'] = dense_backward(
        d_fc1, cache['flat'], params['fc1_weights']
    )

    # Flatten -> Max Pooling
    d_flat = d_fc1.reshape(cache['pool1'].shape)

    # Max Pooling
    d_pool1 = max_pooling_backward(d_flat, cache['conv1'])

    # Convolutional Layer
    d_conv1, grads['conv1_kernel'] = conv2d_backward(d_pool1, X, params['conv1_kernel'])

    return grads
```

```python
def evaluate_accuracy(X, y, params):
    """
    Evaluate the accuracy of the model on a dataset.
    
    Parameters:
        X: Data (n_samples, H, W).
        y: Labels (n_samples,).
        params: Dictionary of weights and biases.
    
    Returns:
        accuracy: Accuracy as a proportion.
    """
    correct = 0
    for i, X_sample in enumerate(X):
        # Forward pass
        conv1 = relu(conv2d(X_sample, params['conv1_kernel']))
        pool1 = max_pooling(conv1)
        flat = flatten(pool1)
        fc1 = relu(dense(flat, params['fc1_weights'], params['fc1_biases']))
        output = softmax(dense(fc1, params['fc2_weights'], params['fc2_biases']))

        # Predicted label
        pred = np.argmax(output)

        # Check if correct
        if pred == y[i]:
            correct += 1

    # Calculate accuracy
    return correct / len(y)
```

```python
def train(X_train, y_train, X_val, y_val, params, epochs, learning_rate):
    """
    Train the CNN using gradient descent and evaluate on a validation set.
    
    Parameters:
        X_train: Training data (n_samples, H, W).
        y_train: Labels (n_samples,).
        X_val: Validation data (n_samples, H, W).
        y_val: Validation labels (n_samples,).
        params: Dictionary of weights and biases.
        epochs: Number of training epochs.
        learning_rate: Step size for gradient updates.
    """
    for epoch in range(epochs):
        epoch_loss = 0
        
        # Training Loop
        for X, y in zip(X_train, y_train):
            # Forward pass
            cache = {}
            conv1 = relu(conv2d(X, params['conv1_kernel']))
            cache['conv1'] = conv1

            pool1 = max_pooling(conv1)
            cache['pool1'] = pool1

            flat = flatten(pool1)
            cache['flat'] = flat

            fc1 = relu(dense(flat, params['fc1_weights'], params['fc1_biases']))
            cache['fc1'] = fc1

            output = softmax(dense(fc1, params['fc2_weights'], params['fc2_biases']))

            # Loss (Cross-Entropy)
            loss = -np.log(output[y])
            epoch_loss += loss

            # Backward pass
            grads = backprop(X, y, output, params, cache)

            # Update parameters
            for key in params:
                params[key] -= learning_rate * grads[key]

        # Calculate Validation Accuracy
        val_accuracy = evaluate_accuracy(X_val, y_val, params)

        # Print Epoch Results
        print(f"Epoch {epoch+1}, Loss: {epoch_loss / len(X_train):.4f}, Validation Accuracy: {val_accuracy * 100:.2f}%")
```

```python
def initialize_parameters():
    """
    Initialize weights and biases for the CNN.
    """
    params = {
        'conv1_kernel': np.random.randn(3, 3) * 0.01,
        'fc1_weights': np.random.randn(169, 64) * 0.01,  # Adjust input size
        'fc1_biases': np.zeros(64),
        'fc2_weights': np.random.randn(64, 10) * 0.01,
        'fc2_biases': np.zeros(10)
    }
    return params
```

```python

# Initialize parameters
params = initialize_parameters()

# Train the model
train(X_train, y_train, params, epochs=10, learning_rate=0.01)
```

## Let's try a one hidden layer NN 

```python
def make_one_versus_all_labels(y, num_classes):
    """
    Converts the given labels into a one-vs-all format for multi-class classification.

    Input:
    - y: Array of shape (n_samples,) containing the original class labels.
    - num_classes: Integer representing the total number of classes.

    Process:
    - Creates a binary label array where each row corresponds to one sample, and the columns
      represent one-vs-all labels for each class.

    Output:
    - Returns a label array of shape (n_samples, num_classes) with 0 for non-class columns
      and 1 for the true class column.
    """
    
    OVA = np.zeros((y.size,num_classes))

    for i,ex in enumerate(y):
        OVA[i,int(ex)] = 1
    
    return OVA
```

```python
class NN_1Layer:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.01, lambda_reg=0.01, class_weights=None):
        np.random.seed(42)  # For reproducibility
        self.H_in = input_size
        self.H_out = output_size
        self.lambda_reg = lambda_reg
        self.learning_rate = learning_rate
        self.class_weights = np.array(class_weights) if class_weights is not None else np.ones(output_size)
        
        self.W1 = np.random.randn(input_size, hidden_size) * np.sqrt(2 / input_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size) * np.sqrt(2 / hidden_size)
        self.b2 = np.zeros((1, output_size))

    #activation function
    def sigmoid(self,x):
        return 1 / (1 + np.exp(-x))

    # Derivative of the sigmoid function
    def sigmoid_derivative(self,x):
        return self.sigmoid(x) * (1 - self.sigmoid(x))
        
    # Forward pass           
    def forward_propagation(self, X):
        Z1 = np.dot(X, self.W1) + self.b1 # Linear step for hidden layer
        A1 = self.sigmoid(Z1)  # Activation for hidden layer
        Z2 = np.dot(A1, self.W2) + self.b2 # Linear step for output layer 
        A2 = self.sigmoid(Z2) # Activation for hidden layer
        return Z1, A1, Z2, A2
        
    # Compute loss (Binary Cross-Entropy)
    def compute_loss(self, Y, A):
        m = Y.shape[0]  # Number of samples
        # Apply class weights
        weight_matrix = Y * self.class_weights  # Weight only the positive examples
        loss = -np.sum(weight_matrix * np.log(A) + (1 - Y) * np.log(1 - A)) / m
        
        # Add L2 regularization
        l1_loss = (self.lambda_reg / (2 * m)) * (np.sum(np.abs(self.W1)) + np.sum(np.abs(self.W2)))
        
        return loss + l1_loss

    def backward_propagation(self, X, Y, Z1, A1, Z2, A2):
        m = X.shape[0]  # Number of samples
        dZ2 = (A2 - Y)  # Gradient of loss w.r.t. Z2
        dW2 = (np.dot(A1.T, dZ2) / m) + (self.lambda_reg / m) * np.sign(self.W2) # Gradient of W2
        db2 = np.sum(dZ2, axis=0, keepdims=True) / m  # Gradient of b2
        dA1 = np.dot(dZ2, self.W2.T)  # Gradient of A1
        dZ1 = dA1 * self.sigmoid_derivative(Z1)  # Gradient of Z1
        dW1 = (np.dot(X.T, dZ1) / m) + (self.lambda_reg / m) * np.sign(self.W1) # Gradient of W1
        db1 = np.sum(dZ1, axis=0, keepdims=True) / m  # Gradient of b1
        return dW1, db1, dW2, db2

    def update_parameters(self,dW1, db1, dW2, db2):
        self.W1 -= self.learning_rate * dW1
        self.b1 -= self.learning_rate * db1
        self.W2 -= self.learning_rate * dW2
        self.b2 -= self.learning_rate * db2

    # Main training function
    def train_neural_network(self, X_train, Y_train, X_valid, Y_valid, epochs=1000):

        self.loss_train = np.zeros(epochs)
        self.loss_valid = np.zeros(epochs)

        self.acc_train = np.zeros(epochs)
        self.acc_valid = np.zeros(epochs)
               
        for epoch in range(epochs):
            # Forward propagation
            Z1_train, A1_train, Z2_train, A2_train = self.forward_propagation(X_train)
            Z1_valid, A1_valid, Z2_valid, A2_valid = self.forward_propagation(X_valid)
            
            # Compute loss
            self.loss_train[epoch] = self.compute_loss(Y_train, A2_train)
            self.loss_valid[epoch] = self.compute_loss(Y_valid, A2_valid)

            self.acc_train[epoch] = np.mean(Y_train.argmax(axis=1) == A2_train.argmax(axis=1))
            self.acc_valid[epoch] = np.mean(Y_valid.argmax(axis=1) == A2_valid.argmax(axis=1))
            
            # Backward propagation
            dW1, db1, dW2, db2 = self.backward_propagation(X_train, Y_train, Z1_train, A1_train, Z2_train, A2_train)
            
            # Update parameters
            self.update_parameters(dW1, db1, dW2, db2)

            
            # Print loss every 10 epochs
            # if epoch%10==0:
            print(f"Epoch {epoch}, Loss_train: {self.loss_train[epoch]:.4f}, Loss_valid: {self.loss_valid[epoch]:.4f}, acc_train: {self.acc_train[epoch]:.2f}, acc_valid: {self.acc_valid[epoch]:.2f}")
            
```

```python
class NeuralNetworkSingleOutput:
    def __init__(self, input_size, hidden_size, learning_rate=0.01, lambda_reg=0.01):
        """
        Initialize the neural network with one hidden layer and single output.
        """
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.learning_rate = learning_rate
        self.lambda_reg = lambda_reg

        # Initialize weights and biases
        np.random.seed(42)
        self.W1 = np.random.randn(input_size, hidden_size) * np.sqrt(2 / input_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, 1) * np.sqrt(2 / hidden_size)
        self.b2 = np.zeros((1, 1))
    
    @staticmethod
    def relu(x):
        """ReLU activation function."""
        return np.maximum(0, x)

    @staticmethod
    def relu_derivative(x):
        """Derivative of ReLU activation."""
        return (x > 0).astype(float)
    
    def forward_propagation(self, X):
        """
        Perform forward propagation through the network.
        """
        self.Z1 = np.dot(X, self.W1) + self.b1  # Hidden layer linear step
        self.A1 = self.relu(self.Z1)           # Hidden layer activation
        self.Z2 = np.dot(self.A1, self.W2) + self.b2  # Output layer linear step
        return self.Z2  # Raw output

    def compute_loss(self, Y, Z2):
        """
        Compute the mean squared error (MSE) loss with L2 regularization.
        """
        m = Y.shape[0]
        # Mean Squared Error
        loss = np.mean((Z2 - Y) ** 2)
        # L2 regularization
        l2_loss = (self.lambda_reg / (2 * m)) * (np.sum(np.square(self.W1)) + np.sum(np.square(self.W2)))
        return loss + l2_loss

    def backward_propagation(self, X, Y, Z2):
        """
        Perform backward propagation and compute gradients.
        """
        m = X.shape[0]
        # Gradients for output layer
        dZ2 = 2 * (Z2 - Y) / m  # Gradient of loss w.r.t. Z2
        dW2 = (np.dot(self.A1.T, dZ2) / m) + (self.lambda_reg / m) * self.W2
        db2 = np.sum(dZ2, axis=0, keepdims=True) / m

        # Gradients for hidden layer
        dA1 = np.dot(dZ2, self.W2.T)
        dZ1 = dA1 * self.relu_derivative(self.Z1)
        dW1 = (np.dot(X.T, dZ1) / m) + (self.lambda_reg / m) * self.W1
        db1 = np.sum(dZ1, axis=0, keepdims=True) / m

        return dW1, db1, dW2, db2

    def update_parameters(self, dW1, db1, dW2, db2):
        """
        Update weights and biases using gradients.
        """
        self.W1 -= self.learning_rate * dW1
        self.b1 -= self.learning_rate * db1
        self.W2 -= self.learning_rate * dW2
        self.b2 -= self.learning_rate * db2

    def train(self, X_train, Y_train, X_valid, Y_valid, epochs=1000):
        """
        Train the network using the provided data.
        """
        self.loss_train = np.zeros(epochs)
        # self.loss_valid = np.zeros(epochs)
        
        for epoch in range(epochs):
            # Forward propagation
            Z2_train = self.forward_propagation(X_train)
            # Z2_valid = self.forward_propagation(X_valid)
            
            # Compute loss
            self.loss_train[epoch] = self.compute_loss(Y_train, Z2_train)
            # self.loss_valid[epoch] = self.compute_loss(Y_valid, Z2_valid)
            
            # Backward propagation
            dW1, db1, dW2, db2 = self.backward_propagation(X_train, Y_train, Z2_train)
            
            # Update parameters
            self.update_parameters(dW1, db1, dW2, db2)
            
            # Print loss every 100 epochs
            # if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss_train: {self.loss_train[epoch]:.4f}")#, Loss_valid: {self.loss_valid[epoch]:.4f}")

    def predict(self, X):
        """
        Predict the class for given input data by rounding the output.
        """
        Z2 = self.forward_propagation(X)
        return np.round(Z2).astype(int).clip(0, 3)  # Clip predictions to valid class range [0, 3]
```
